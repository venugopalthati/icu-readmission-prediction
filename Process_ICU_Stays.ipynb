```python
#Order stays by stayId

stays.sort_values(by=['icustay_id'])

errorcount = 0
total_stays = len(stays)
print('total_stays == ', total_stays)
max_value = 48
event_channels = 59
demograhic_channels = 14
icd9_channels = 17
chart_data = np.zeros((total_stays, max_value, event_channels))
demographic_data = np.zeros((total_stays, demograhic_channels))
icd9_data = np.zeros((total_stays, icd9_channels))

y = np.zeros((total_stays))
stays['intime'] = pd.to_datetime(stays['intime'])
stays['outtime'] = pd.to_datetime(stays['outtime'])
errorcount = 0
#stays = stays[stays['icustay_id'] == 224504]
#for stay_index, row in stays_1.iterrows():
    #stay_index  = 0
#do_parallel_stuff_on_dataframe(stays_1, num_cores)
for stay_index, row in stays.iterrows():
        try:
            print('processing index', stay_index)

            #print(row['los'])
            intime = row['intime']
            #print(len(events[events['icustay_id'] == row['icustay_id']]))

            #print(filteredevents)

            temp_events = events[(events['icustay_id'] ==  row['icustay_id']) & \
                                 (events['charttime'] >= row['intime']) & (events['charttime'] <= row['outtime'])].copy()
            #print(temp_events)
            metadata = temp_events[['charttime', 'icustay_id']].sort_values(by=['charttime', 'icustay_id'])\
                            .drop_duplicates(keep='first').set_index('charttime')

            #print(metadata.columns.tolist())

            timeseries = temp_events[['charttime', 'event_name','event_val']]\
                                .sort_values(by=['charttime', 'event_name','event_val'], axis=0)\
                                    .drop_duplicates(subset=['charttime','event_name'], keep='last')

            #print(timeseries)
            timeseries['charttime'] = pd.to_datetime(timeseries['charttime'])
            timeseries['event_time'] = (timeseries['charttime'] - intime).astype('timedelta64[h]').astype('int64')


            if len(timeseries['event_time'].unique()) > 0: #handling no events at all
                max_hours = int(timeseries['event_time'].max())
                expected_entries = 47
                starthour = 0
                if max_hours > expected_entries:
                    starthour = max_hours - expected_entries
                startIndex = 0
                for i in range(starthour, max_hours+1):
                    events_by_hour = timeseries[timeseries['event_time'] == i].copy()
                    if len(events_by_hour) > 0:  #We may need to handle missing entries in between
                        for index1, row1 in events_by_hour.iterrows():
                                #print(row['event_name'])
                                #print(chart_event_index[row['event_name']])
                                #print(row1['event_name'], row1['event_val'])
                                if row1['event_name'] == 'Glucose' and not str(row1['event_val']).isnumeric():
                                    continue;
                                if row1['event_val'] == 'CS 131' or row1['event_val'] == 'CS 223'\
                                or row1['event_val'] == 'cs 148' or row1['event_val'] == 'cs 130':
                                    continue;
                                if row1['event_val'] == 'CS 164':
                                    continue;
                                if row1['event_val'] == 'CS 178':
                                    continue;
                                if row1['event_name'] not in chart_event_index:
                                    #print('skipping',row['icustay_id'],row1['event_name'],row1['event_val'])
                                    continue;
                                chart_data[stay_index][startIndex][chart_event_index[row1['event_name']]] = row1['event_val']
                    startIndex = startIndex + 1

                lastIndex = startIndex
                if max_hours < expected_entries:
                    #print('Less hours', max_hours+1, expected_entries+1)
                    for i in range(max_hours+1, expected_entries+1):
                            chart_data[stay_index, startIndex,:] = chart_data[stay_index, lastIndex, :]
                            startIndex = startIndex + 1


            icd9groups_m = icd9groups.loc[icd9groups['icustay_id'] ==  row['icustay_id'],('group_name', 'group_val')].copy()

            #print(icd9groups_m)
            #print(icd9groups_m.columns.tolist())
            for index2, row2 in icd9groups_m.iterrows():
                  icd9_data[stay_index][icd9_index[row2['group_name']]] = row2['group_val']

            demographic_data[stay_index][demographic_index[row['gender']]] = 1
            demographic_data[stay_index][demographic_index[row['ethnicity']]] = 1
            demographic_data[stay_index][demographic_index[row['insurance']]] = 1
            demographic_data[stay_index][demographic_index['AGE']] = row['age']

            y[stay_index] = row['y']

        except Exception as e:
            errorcount = errorcount + 1
            print("in exception", errorcount)
            #print(f"{type(e).__name__} at line {e.__traceback__.tb_lineno} of {__file__}: {e}")
            continue
            
    
```
